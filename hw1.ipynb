给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

我们遍历收益数组，并维护一个当前连续子数组的和。如果我们当前的累积收益变成负数，说明之前的走势只会拖累未来的收益，因此我们应该抛弃之前的选择，从下一天重新开始计算潜在的买入点。在遍历过程中，不断更新全局最大和以及对应的买入和卖出日期。

````python
# add your code here
def max_profit_days(profits):
    if not profits:
        return 0, None, None
    max_sum = -float('inf')
    current_sum = 0
    best_buy = 0
    best_sell = 0
    temp_buy_day = 0 
    for i in range(len(profits)):
        if current_sum <= 0:
            current_sum = profits[i]
            temp_buy_day = i
        else:
            current_sum += profits[i]
        if current_sum > max_sum:
            max_sum = current_sum
            best_buy = temp_buy_day
            best_sell = i + 1
    return max_sum, best_buy + 1, best_sell + 1
# your algorithm time complexity is:O(n)
````

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

我们可以使用二分查找法，首先取中点，检查A[mid]与其相邻元素A[mid-1]和A[mid+1]的关系。之后判断状态，如果A[mid]比两侧都小，那么A[mid]就是局部最小值。如果A[mid-1] < A[mid]，说明左侧处于下降阶段，左侧区间一定存在局部最小值。如果A[mid+1] < A[mid]，同理，右侧区间一定存在局部最小值，每次丢弃一半不符合条件的区间。

````python
# add your code here
def find_one_local_minimum(A):
    n = len(A)
    if n == 1: return A[0]
    low = 0
    high = n - 1
    
    while low <= high:
        mid = (low + high) // 2
        prev = A[mid - 1] if mid > 0 else float('inf')
        nxt = A[mid + 1] if mid < n - 1 else float('inf')
        if A[mid] < prev and A[mid] < nxt:
            return A[mid]
        elif mid > 0 and A[mid-1] < A[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return None
# your algorithm time complexity is: O(logn)
````

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

首先创建一个空栈stack用于存放存活的星球。逐个处理数组A中的星球p。如果 p > 0（向右），它暂时不会与左侧的任何星球碰撞，直接入栈。如果 p < 0（向左），它可能会与栈顶向右移动的星球发生碰撞。只要栈不为空，且栈顶元素 top > 0，若 |top| < |p|：栈顶星球质量小，被摧毁（弹出栈顶），继续检查下一个栈顶。若 |top| > |p|：当前星球 p 质量小，被摧毁，停止检查。若|top| == |p|，根据规则，向左运行的 p 消失，向右的 top 保留，停止检查。如果 p 经过一系列碰撞后没有被摧毁（即栈为空，或者栈顶也是向左运行的星球），则将 p 入栈。遍历结束后，栈中剩余的元素即为最终存续的星球。

````python
# add your code here
def solve_planet_collision(A):
    stack = []
    for p in A:
        alive = True
        while alive and stack and stack[-1] > 0 and p < 0:
            top = stack[-1]
            if abs(top) < abs(p):
                stack.pop()
                continue
            elif abs(top) > abs(p):
                alive = False
            else:
                alive = False
            break
        if alive:
            stack.append(p)
    return stack
# your algorithm time complexity is:O(n)
````

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

0-1背包问题，用动态规划。设dp[i][w]为在前i件物品中选择、且推车当前最大承重为w时能获得的最大价值。

对于第 $i$ 件物品（重量为 $w_i$，价值为 $v_i$）：不选第 $i$ 件物品：$dp[i][w] = dp[i-1][w]$选择第 $i$ 件物品（前提是当前容量 $w \ge w_i$）：$dp[i][w] = dp[i-1][w - w_i] + v_i$

综上所述，状态转移方程为：$dp[i][w] = max(dp[i-1][w], dp[i-1][w - w_i] + v_i)$

````python
# add your code here
def knapsack_01(weights, values, capacity):
    n = len(weights)
    dp = [0] * (capacity + 1)
    for i in range(n):
        for w in range(capacity, weights[i] - 1, -1):
            dp[w] = max(dp[w], dp[w - weights[i]] + values[i])
    return dp[capacity]
# your algorithm time complexity is: O(N*W)
````

idea:

当BST中有两个节点被错误交换后，其中序遍历序列中会出现逆序对。如果我们找到第一个逆序对(prev, curr)，其中prev > curr，那么第一个错误节点一定是prev。如果我们找到第二个逆序对(prev, curr)，那么第二个错误节点一定是curr。如果全序列只有一个逆序对，那么这两个节点就是我们要找的prev和curr。先对二叉树进行中序遍历，在遍历过程中，记录上一个访问的节点prev_node。比较 prev_node.val 和 curr.val。如果发生逆序，记录这两个节点。遍历结束后，交换这两个错误节点的值。

````python
# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        self.first = None
        self.second = None
        self.prev = TreeNode(float('-inf')) # 初始化为无穷小
        def inorder(node):
            if not node:
                return
            inorder(node.left)
            if self.prev.val > node.val:
                self.second = node
                if not self.first:
                    self.first = self.prev
            self.prev = node 
            inorder(node.right)
        inorder(root)
        if self.first and self.second:
            self.first.val, self.second.val = self.second.val, self.first.val
# your algorithm time complexity is:O(N)
````

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:

这道题使用深度优先遍历，假设当前路径形成的数字为current_sum，当我们移动到子节点 node 时，新的路径值new_sum = current_sum * 10 + node.val。当我们到达一个叶子节点时，当前的路径值就是一个完整的数字，将其累加到总和中。从根节点出发，初始路径和为 0。每进入一层，将上一层的和乘以 10 并加上当前节点的值。如果是叶子节点，返回当前和。如果不是叶子节点，递归计算左子树和右子树的和，并返回它们的总和。

````python
# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def sumNumbers(root: TreeNode) -> int:
    def dfs(node, current_sum):
        if not node:
            return 0
        current_sum = current_sum * 10 + node.val
        # 如果是叶子节点，直接返回当前路径和
        if not node.left and not node.right:
            return current_sum
        # 否则，递归计算左右子树的和
        return dfs(node.left, current_sum) + dfs(node.right, current_sum)

    return dfs(root, 0)
# your algorithm time complexity is: O(N)
````
